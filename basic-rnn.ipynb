```python
# 4. Generate text using the trained model
def generate_text(model, start_string, gen_length=100):
    # """Generates new text starting from the start_string."""
    print(start_string)
    # Convert start_string to numbers (vectorize)
    input_eval = [char2idx[char] for char in start_string]
    input_eval = np.array(input_eval).reshape(1, -1)  # Shape (1, seq_length)
    
    # Empty string to collect generated text
    generated_text = start_string
    
    for _ in range(gen_length):
        
        # Get predictions from the model
        predictions = model.predict(input_eval, verbose=0)
        
        # Sample the next character from the probability distribution
        predicted_idx = np.argmax(predictions)
        
        # Convert number back to character
        next_char = idx2char[predicted_idx]
        
        # Add next character to the generated text
        generated_text += next_char
        
        # Update the input for the next prediction
        input_eval = np.append(input_eval[:, 1:], [[predicted_idx]], axis=1)
    
    return generated_text
```
