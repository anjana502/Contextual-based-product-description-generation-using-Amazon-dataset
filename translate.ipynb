```python
import random
import re
from translate import Translator
```

```python
input_string = "Design a [noun] for [adverb] [verb-ing form] [job people or family] that is [adjective], and [adverb of frequency] makes them [verb], and makes them feel [emotion]"
```

```python
# Define a function to replace words based on word type
def replace_words(match):
    word_type = match.group(1)
    if word_type == "noun":
        return random.choice(nouns)
    elif word_type == "adverb":
        return random.choice(adverbs)
    elif word_type == "verb-ing form":
        return random.choice(verb_ing_forms)
    elif word_type == "job people or family":
        return random.choice(jobs)
    elif word_type == "adjective":
        return random.choice(adjectives)
    elif word_type == "adverb of frequency":
        return random.choice(adverbs_of_frequency)
    elif word_type == "verb":
        return random.choice(verbs)
    elif word_type == "emotion":
        return random.choice(emotions)
    else:
        return match.group(0)  # Return the original word if type not recognized
```

```python
# Lists of words for each word type
nouns = ["box", "train", "house", "book", "flower", "tree", "car", "city", "computer", "pen", "traffic light", "light bulb"]
adverbs = ["quickly", "softly", "carefully", "slowly", "happily", "loudly", "gently", "recently"]
verb_ing_forms = ["running", "swimming", "writing", "dancing", "singing", "jumping", "graduating", "working"]
jobs = ["teacher", "doctor", "engineer", "chef", "artist", "lawyer", "scientist", "musician", "parents", "couple", "grandfathers", "grandmothers", "brothers", "sisters", "friends", "colleagues", "boss"]
adjectives = ["happy", "colorful", "amazing", "creative", "fun", "bright", "quiet", "tasty", "sunny"]
adverbs_of_frequency = ["always", "often", "sometimes", "rarely", "never", "usually", "occasionally"]
verbs = ["dance", "read", "play", "sing", "write", "run", "swim", "eat", "sleep", "talk", "change", "cook"]
emotions = ["happy", "sad", "excited", "relaxed", "angry", "joyful", "nervous", "grateful", "surprised"]
```

```python
# Split the input string into phrases
phrases = input_string.split(',')

# Initialize the output string
output_string = ""

# Create a list to hold optional phrases
optional_phrases = [
    "for [adverb] [verb-ing form] [job people or family]",
    "and [adverb of frequency] makes them [verb]",
    "and makes them feel [emotion]"
]

# Randomly include optional phrases based on word presence
selected_optional_phrases = []  # Keep track of selected optional phrases
```

```python
# Randomly include optional phrases based on word presence
for phrase in phrases:
    phrase = phrase.strip()
    if "[noun]" in phrase or "[adjective]" in phrase:
        # For mandatory phrases, always replace words randomly
        phrase = re.sub(r'\[([^\]]+)\]', replace_words, phrase)
    else:
        # Check if the specified mandatory word types are present
        mandatory_word_types = ["[job people or family]", "[verb]", "[emotion]"]
        mandatory_present = all(word_type in phrase for word_type in mandatory_word_types)

        # Randomly decide whether to include the optional phrase
        include_optional = random.choice([True, False]) if not mandatory_present else True

        if include_optional:
            # If at least one word type is present, include a non-repeated optional phrase
            unused_optional_phrases = [phrase for phrase in optional_phrases if phrase not in selected_optional_phrases]
            if unused_optional_phrases:
                selected_phrase = random.choice(unused_optional_phrases)
                selected_optional_phrases.append(selected_phrase)
                selected_phrase = re.sub(r'\[([^\]]+)\]', replace_words, selected_phrase)
                phrase = selected_phrase
            else:
                # If all optional phrases have been used, exclude the phrase
                phrase = ""
        else:
            # Otherwise, exclude the phrase
            phrase = ""
    if phrase:
        output_string += phrase + ", "
```

```python
# Remove trailing comma and space
output_string = output_string.rstrip(', ')
# target_language = input("Enter the language you want to generate in 2 small letters code: ")

# Print the modified string
translator = Translator(to_lang='en')
translated_output = translator.translate(output_string)

# Print the translated output
print(translated_output)
```
